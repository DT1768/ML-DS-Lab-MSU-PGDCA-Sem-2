```python
#Write a python code for linear regression
#write a python code for binomial distribution
#  Example
#  Syntax Description
#  Code
#  Input
#  Output
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../input/linearregressiondata/Data Linear Regression.csv',header=0)

n = df.shape[0]

sumy = df["Life Expectancy"].sum()
sumx = df["Year"].sum()
temp = df["Year"]**2
sumx2 = temp.sum()
temp1 = df["Life Expectancy"]*df["Year"]
sumxy = temp1.sum()

a = [[sumx,n],[sumx2,sumx]]
A = np.array(a)

b = [[sumy],[sumxy]]
B = np.array(b)

X = np.linalg.inv(A).dot(B)

m = np.linspace(1990,2025)
n = X[0]*m + X[1]

print(X)
plt.plot(m,n,'-r')
plt.scatter(df["Year"],df["Life Expectancy"])
plt.title("Linear Regression")
plt.grid()
plt.show()

#print(n)
#print(sumx)
#print(sumy)
#print(sumx2)
#print(sumxy)
#print(X)
```
